```python
import pandas as pd
import os

# Set the directory containing the CSV files
directory = "../structured_data/"

# Initialize an empty dictionary to store the DataFrames
data_dict = {}

# Loop through all CSV files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        # Get the filename without the extension
        name = os.path.splitext(filename)[0]
        filepath = os.path.join(directory, filename)
        # Read the CSV file into a Pandas DataFrame
        df = pd.read_csv(filepath)
        # Append the DataFrame to the dictionary using the filename as the key
        data_dict[name] = df

# Print the dictionary to verify the contents
for key, dataFrame in data_dict.items():
    print(f"Table: {key}, St. vrstic: {len(dataFrame)}")
    display(dataFrame.info())
```

```python
for column in data_dict:
    print(column)
```

```python
data_dict["delovno_prebivalstvo"].columns = ['num', 'obcina', 'leto', 'izobrazba', 'spol']
data_dict["delovno_prebivalstvo"]
```

```python
data_dict["delovno_prebivalstvo"].loc[(data_dict["delovno_prebivalstvo"]['obcina'] == "Bled") &
                                     (data_dict["delovno_prebivalstvo"]['leto'] >= 2018) &
                                     (data_dict["delovno_prebivalstvo"]['izobrazba'] == "Osnovno�olska ali manj")]
```

```python
# Podatke o brezposelnosti imamo od 2018 naprej.

obcine = data_dict["delovno_prebivalstvo"]["obcina"].unique()

for obcina in obcine:
    
    izobrazbe = data_dict["delovno_prebivalstvo"]["izobrazba"].unique()
    for izobrazba in izobrazbe:
        
        data_dict["delovno_prebivalstvo"] =\
            data_dict["delovno_prebivalstvo"].loc[(data_dict["delovno_prebivalstvo"]['leto'] >= 2018)]
```

```python
data_dict["po-obcinah_izobrazba"].rename({
    "Občina":"občina",
    "OŠ ali manj":"OŠ",
    "srednje poklicno izobraževanje":"poklicna",
    "strokovno, splošno izobraževanje":"strokovna",
    "izobraževanje prve, druge, tretje stopnje":"faks"}, axis='columns', inplace=True);
data_dict["po-obcinah_izobrazba"]
```

```python
data_dict["po-obcinah_izobrazba"].dropna(axis='rows', how='any', inplace=True)

for col in data_dict["po-obcinah_izobrazba"].columns:
    data_dict["po-obcinah_izobrazba"]\
        .drop(data_dict["po-obcinah_izobrazba"]\
        .loc[data_dict["po-obcinah_izobrazba"][col] == "-"].index, inplace=True)

data_dict["po-obcinah_izobrazba"] = data_dict["po-obcinah_izobrazba"].astype({
    'OŠ':'float',
    'poklicna':'float',
    'strokovna':'float',
    'faks':'float'})
```

```python
# Treba je agregirati podatke o bresposelnosti na letno raven z izračunom povprečja mesecev za posamezno leto.
# procenti-izobrazba.txt

data_dict["po-obcinah_izobrazba"].groupby(["občina", "month"]).agg({
    "OŠ":"mean",
    "poklicna":"mean",
    "strokovna":"mean",
    "faks":"mean"})

data_dict["po-obcinah_izobrazba"] = data_dict["po-obcinah_izobrazba"].groupby(['year', 'občina'])[['OŠ', 'poklicna', 'strokovna', 'faks', 'Skupaj']].agg('mean')
data_dict["po-obcinah_izobrazba"].reset_index(inplace=True)
data_dict["po-obcinah_izobrazba"]
```
